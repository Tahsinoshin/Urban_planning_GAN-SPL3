```python
import cv2
import shutil
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Load the uploaded image
image = cv2.imread("B_4_Groundplan_2011.jpg")

# Get the dimensions of the image
height, width, _ = image.shape

# Define the size of the sub-images
sub_image_size = 512

sub_images = []
for y in range(0, height, sub_image_size):
    for x in range(0, width, sub_image_size):
        sub_image = image[y:y+sub_image_size, x:x+sub_image_size]
        sub_images.append(sub_image)



# Display the sub-images
# Adjust the figure size to make the images larger
plt.figure(figsize=(5, 15))  # Adjust the width and height as needed

# Display the sub-images vertically
for i, sub_image in enumerate(sub_images):
    plt.subplot(len(sub_images), 1, i + 1)  # Create a grid with rows for each sub-image
    plt.imshow(cv2.cvtColor(sub_image, cv2.COLOR_BGR2RGB))

    # Remove the axis to make it look better
    plt.axis('off')

    # Adjust the aspect ratio to make sure the images are not stretched
    plt.gca().set_aspect('equal')

# Show the vertically stacked sub-images
plt.show()



```

```python
compressed_sub_images = [cv2.resize(sub_image, (256, 256)) for sub_image in sub_images]
```

```python
import matplotlib.pyplot as plt

for i, sub_image in enumerate(compressed_sub_images):
    plt.subplot(1, len(compressed_sub_images), i + 1)
    plt.imshow(cv2.cvtColor(sub_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

plt.show()
```

```python
import cv2
import os
import zipfile
from google.colab import files

# Specify the path to the uploaded zip file
# zip_file_path = "MAP.zip"

# Specify the target directory to unzip the contents
target_directory = "/content/drive/MyDrive/extracted_folder/MAP"

# Unzip the contents
# with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
#    zip_ref.extractall(target_directory)

# Now that the files are unzipped, you can use them in your image processing code.

# Directory where the original images are stored
original_images_folder = target_directory  # Use the unzipped directory

# Directory to save processed images
processed_images_folder = "/content/Downloads/processed_images"

# Create the processed images folder
os.makedirs(processed_images_folder, exist_ok=True)

# Function to create sub-images
def create_sub_images(image):
    sub_images = []
    for i in range(0, image.shape[0], 700):
        for j in range(0, image.shape[1], 700):
            sub_image = image[i:i+700, j:j+700]
            sub_images.append(sub_image)
    return sub_images

# Function to compress an image to 256x256 without losing information
def resize_image(image, new_size):
    return cv2.resize(image, (new_size, new_size))

for filename in os.listdir(original_images_folder):
    if filename.endswith((".jpg", ".png", ".jpeg")):  # Provide extensions as a tuple
        image = cv2.imread(os.path.join(original_images_folder, filename))
        original_image_name = os.path.splitext(filename)[0]

        # Create a folder for the original image in the download directory
        original_image_download_folder = os.path.join(processed_images_folder, original_image_name)
        os.makedirs(original_image_download_folder, exist_ok=True)

        # Split the image into sub-images of size 700x700
        sub_images = create_sub_images(image)

        # Compress sub-images to 256x256 without losing information
        for i, sub_image in enumerate(sub_images):
            compressed_image = resize_image(sub_image, 256)

            # Save the compressed image
            sub_image_filename = os.path.join(original_image_download_folder, f"sub_image_{i}.jpg")
            cv2.imwrite(sub_image_filename, compressed_image)

# Zip each folder with the sub-images and provide download links
for folder in os.listdir(processed_images_folder):
    if os.path.isdir(os.path.join(processed_images_folder, folder)):
        shutil.make_archive(
            os.path.join(processed_images_folder, folder), 'zip', os.path.join(processed_images_folder, folder)
        )

# Provide download links for each zip archive
for folder in os.listdir(processed_images_folder):
    if os.path.isdir(os.path.join(processed_images_folder, folder)):
        zip_filename = os.path.join(processed_images_folder, f"{folder}.zip")
        files.download(zip_filename)
```

```python
import cv2
import os
from google.colab import files

# Specify the path to the uploaded zip file
# zip_file_path = "MAP.zip"

# Specify the target directory to unzip the contents
target_directory = "/content/drive/MyDrive/extracted_folder/MAP"

# Unzip the contents
# with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
#    zip_ref.extractall(target_directory)

# Now that the files are unzipped, you can use them in your image processing code.

# Directory where the original images are stored
original_images_folder = target_directory  # Use the unzipped directory

# Load the first image in the folder
first_image = None
for filename in os.listdir(original_images_folder):
    if filename.endswith((".jpg", ".png", ".jpeg")):  # Provide extensions as a tuple
        image = cv2.imread(os.path.join(original_images_folder, filename))
        first_image = image
        break

if first_image is not None:
    # Create a folder for the original image in the download directory
    original_image_download_folder = os.path.join("/content/Downloads", "first_image")
    os.makedirs(original_image_download_folder, exist_ok=True)

    # Split the image into sub-images of size 700x700
    sub_images = create_sub_images(first_image)

    # Save all the sub-images as "image1.jpg", "image2.jpg", ...
    for i, sub_image in enumerate(sub_images):
        sub_image_filename = os.path.join(original_image_download_folder, f"image{i + 1}.jpg")
        cv2.imwrite(sub_image_filename, sub_image)

    # Provide download links for all sub-images
    for i in range(len(sub_images)):
        sub_image_filename = os.path.join(original_image_download_folder, f"image{i + 1}.jpg")
        files.download(sub_image_filename)
```

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload an image from your local PC
uploaded = files.upload()

# Check if any files were uploaded
if uploaded:
    # Get the first uploaded file
    file_name = list(uploaded.keys())[0]

    # Read the uploaded image
    image = cv2.imdecode(np.frombuffer(uploaded[file_name], np.uint8), -1)

    # Display the original image
    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis('off')
    plt.show()

    # Resize the image to 256x256 without cropping
    resized_image = cv2.resize(image, (256,256),  interpolation = cv2.INTER_AREA)

    # Display the resized image
    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
    plt.title("Resized Image (256x256)")
    plt.axis('off')
    plt.show()
else:
    print("No files were uploaded.")
```
