```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Define the dimension of the random noise vector
noise_dim = 100  # You can adjust this as needed

# Build a Pix2Pix Generator
def build_generator():
    model = keras.Sequential([
        keras.layers.Dense(256, input_dim=noise_dim, activation='relu'),
        keras.layers.Reshape((16, 16, 1)),
        keras.layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same', activation='relu'),
        keras.layers.Conv2DTranspose(64, (4, 4), strides=(2, 2), padding='same', activation='relu'),
        keras.layers.Conv2D(1, (4, 4), padding='same', activation='sigmoid')
    ])
    return model

generator = build_generator()

# Generate Images from Random Noise
num_images = 10
random_noise = np.random.normal(0, 1, size=(num_images, noise_dim))

generated_images = generator.predict(random_noise)

# Plot Generated Images
def plot_generated_images(images):
    plt.figure(figsize=(12, 6))
    for i in range(num_images):
        plt.subplot(2, num_images // 2, i + 1)
        plt.imshow(images[i, :, :, 0], cmap='gray')
        plt.axis('off')
    plt.show()

# Plot the generated images
plot_generated_images(generated_images)
```
